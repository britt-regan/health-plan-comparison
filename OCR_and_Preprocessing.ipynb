```python
import PyPDF2
import io
import pytesseract
from nltk.corpus import stopwords
import re
```

```python
#Open and OCR PDF files

pdf_file = open('Devoted-PRIME-(HMO)-EOC-H1290-037-001-ENG.pdf', 'rb') #Devoted 4
#pdf_file = open('AARP_AAFL23LP0051092_003.pdf', 'rb') #AARP 1
#pdf_file = open('AARP_AAFL23RP0051470_003.pdf', 'rb') #AARP 2
#pdf_file = open('AvMed_ce545-082022_c-eoc-choice-miami-dade-001.pdf', 'rb') #AvMed 1
#pdf_file = open('AvMed_ce547-082022_c-eoc_circle-miami-dade-23.pdf', 'rb') #AvMed 2
#pdf_file = open('AvMed_ce549-082022_c-eoc_access-miami-dade-025.pdf', 'rb') #AvMed 3
#pdf_file = open('CarePlus_H1019121000EOC23.pdf', 'rb') #CarePlus 1
#pdf_file = open('CarePlus_H1019136000EOC23.pdf', 'rb') #CarePlus 2
#pdf_file = open('Devoted-CORE-Miami-Dade-(HMO)-EOC-H1290-001-ENG.pdf', 'rb') #Devoted 1
#pdf_file = open('Devoted-DUAL-Miami-Dade-(HMO-D-SNP)-EOC-H1290-019-ENG.pdf', 'rb') #Devoted 2
#pdf_file = open('Devoted-ESSENTIALS-Miami-Dade-(HMO)-EOC-H1290-013-ENG.pdf', 'rb') #Devoted 3
pdf_reader = PyPDF2.PdfReader(pdf_file)
print(len(pdf_reader.pages))

#Create empty string to store text
text = ''

#Extract each page and OCR it
for i in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[i]
    page_num = i+1
    
    #Extract text from page using PyPDF2
    extracted_text = page.extract_text()
    
    #Save text to empty string, along with page number
    #text += extracted_text
    text += 'Page_{}: \n'.format(page_num) + extracted_text + '\n\n'
    
    #Print extracted text
    #print(extracted_text)
```

## Text Cleaning & Preprocessing

```python
text = text.lower()
```

```python
#Remove punctuation and special characters
text = re.sub(r'[^\w\s]','', text)
text = re.sub(r'\S*@\S*\s*','', text)
    
#Replace newline characters
text = text.replace('\n',' ')
    
#Replace URLs with "URL"
#text = re.sub(r'http\S+|www.\S+', ' URL', text)
#text = text.replace('http',' ')
        
#Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = text.split()
cleaned_text = [word for word in tokens if word not in stop_words]
text = ' '.join(cleaned_text)
```
